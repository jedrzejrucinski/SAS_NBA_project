```python
dfs = []
for year in years:
   with open(f"allstars/{year}.html") as f:
    page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    starters_E = soup.find(id = "all-star-starters_1")
    starters_W = soup.find(id = "all-star-starters_2")
    starters__E = pd.read_html(str(starters_E))[0]
    starters__E = starters__E.iloc[1:,:]
    starters__W = pd.read_html(str(starters_W))[0]
    starters__W = starters__W.iloc[1:,:]
    starters = pd.concat([starters__E,starters__W],axis = 0)
    starters = starters.reset_index()
    starters = starters.sort_values(2, ascending = False) 
    starters["Year"] = year
    del starters["index"]

    dfs.append(starters)
```

```python
allstar_starters = pd.concat(dfs)
allstar_starters = allstar_starters.reset_index()
del allstar_starters["index"]
allstar_starters.rename({0: 'Position', 1: 'Player', 2: 'Votes'}, axis = 1, inplace = True)
allstar_starters
```

```python
allstar_starters.to_csv("allstar_starters.csv")
```
